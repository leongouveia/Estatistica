    Questão 1: Liste as diferenças entre uma lista, um dicionário, um array numpy e uma series pandas?

# Diferenças :
-    Lista, Caso voce precise de uma coisa ordenada e indexada por posição, pode haver elementos de tipo diferente, coisa que o array não faz.
-   Dicionario, Não é obrigado a ser ordenado e o acesso aos itens e é feito por chaves.
-   Array Numpy, Tem o mesmo tipo de dados, é mais leve em termos de memoria, da para fazer a operações em todos os items de forma que nao será feito um por um.
-  Series pandas, é um array que pode receber qualquer tipo de objeto e tem 1 dimensão
